```python
from pyvis.network import Network
import networkx as nx

# Function to insert line breaks into long labels
def insert_line_breaks(label, max_length=20):
    words = label.split(' ')
    new_label = ''
    line_length = 0
    for word in words:
        if line_length + len(word) + 1 > max_length:
            new_label += '<br>'
            line_length = 0
        new_label += word + ' '
        line_length += len(word) + 1
    return new_label.strip()

# Create a new NetworkX graph
G = nx.DiGraph()

# Add nodes and edges
G.add_node('Android', label='Android')
G.add_node('Sdk', label='Sdk')
G.add_edge('Android', 'Sdk')

# Level 1 nodes
folders = ['add-ons', 'build-tools', 'docs', 'extras', 'licenses', 'patcher', 'platform-tools', 'platforms', 
           'samples', 'skins', 'sources', 'system-images', 'temp', 'tools']
for folder in folders:
    G.add_node(folder, label=insert_line_breaks(folder))
    G.add_edge('Sdk', folder)

# Example of adding subdirectories for platform-tools
platform_tools_files = ['adb', 'api', 'dmtracedump', 'etc1tool', 'fastboot', 'hprof-conv', 'NOTICE.txt', 
                        'package.xml', 'source.properties', 'sqlite3', 'systrace']
for file in platform_tools_files:
    G.add_node(file, label=insert_line_breaks(file))
    G.add_edge('platform-tools', file)

# Example of adding subdirectories for tools
tools_files = ['android', 'ant', 'apps', 'bin', 'lib', 'lib64', 'lint', 'proguard', 'qemu', 'support', 
               'templates', 'traceview', 'uiautomatorviewer']
for file in tools_files:
    G.add_node(file, label=insert_line_breaks(file))
    G.add_edge('tools', file)

# Create a pyvis network
net = Network(notebook=True, directed=True)

# Load the NetworkX graph into the PyVis network
net.from_nx(G)

# Customize the appearance
for node in net.nodes:
    node['title'] = node['label']
    node['label'] = node['label']  # Display the label next to the node
    node['font'] = {'size': 10, 'align': 'left'}  # Align the font to the left of the node

# Set the physics layout to hierarchical
net.set_options("""
var options = {
  "nodes": {
    "font": {
      "align": "left",
      "multi": true
    }
  },
  "layout": {
    "hierarchical": {
      "enabled": true,
      "direction": "LR",
      "sortMethod": "directed"
    }
  },
  "physics": {
    "hierarchicalRepulsion": {
      "nodeDistance": 120
    }
  }
}
""")

# Generate and save the network HTML
net.save_graph("folder_structure.html")

# Custom JavaScript for highlighting paths on click
custom_js = """
<script type="text/javascript">
var container = document.getElementById("mynetwork");
var data = {
  nodes: new vis.DataSet(%s),
  edges: new vis.DataSet(%s)
};
var options = {
  nodes: {
    font: {
      align: "left",
      multi: true
    }
  },
  layout: {
    hierarchical: {
      enabled: true,
      direction: "LR",
      sortMethod: "directed"
    }
  },
  physics: {
    hierarchicalRepulsion: {
      nodeDistance: 120
    }
  }
};
var network = new vis.Network(container, data, options);
network.on("click", function (params) {
    if (params.nodes.length > 0) {
        var clickedNode = params.nodes[0];
        var stack = [clickedNode];
        var highlightedNodes = new Set();
        while (stack.length > 0) {
            var node = stack.pop();
            highlightedNodes.add(node);
            var incomingEdges = data.edges.get({
                filter: function (edge) {
                    return edge.to == node;
                }
            });
            for (var edge of incomingEdges) {
                stack.push(edge.from);
            }
        }
        // Reset all nodes to default color
        var allNodes = data.nodes.get();
        for (var node of allNodes) {
            data.nodes.update({id: node.id, color: 'blue'});
        }
        // Highlight path nodes in red
        for (var nodeId of highlightedNodes) {
            data.nodes.update({id: nodeId, color: 'red'});
        }
    }
});
</script>
""" % (net.get_nodes(), net.get_edges())

# Read the saved HTML file and insert the custom JavaScript
with open("folder_structure.html", "r") as file:
    html_content = file.read()

# Insert the custom JavaScript before the closing body tag
html_content = html_content.replace("</body>", custom_js + "\n</body>")

# Save the modified HTML file
with open("folder_structure.html", "w") as file:
    file.write(html_content)
```
